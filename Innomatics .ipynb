# Haversine distance

```python
R = 6371.0
def haversine_distance(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2]) 
    dlat = lat2 - lat1
    dlon = lon2 - lon1  
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    distance = R * c
    return distance
data['distance'] = data.apply(lambda row: haversine_distance(row['pickup_latitude'], row['pickup_longitude'], row['dropoff_latitude'], row['dropoff_longitude']), axis=1)
median_distance = data['distance'].median()
print(f"The median Haversine distance between pickup and dropoff locations is {median_distance:.2f} kilometers.")
```

```python
max_distance = data['distance'].max()

print(f"The maximum Haversine distance between pickup and dropoff locations is {max_distance:.2f} kilometers.")
```

```python
rides_with_zero_distance = (data['distance'] == 0.0).sum()

print(f"The number of rides with a Haversine distance of 0.0 between pickup and dropoff locations is {rides_with_zero_distance}.")
```

```python
rides_with_zero_distance = data[data['distance'] == 0.0]

mean_fare_for_zero_distance = rides_with_zero_distance['fare_amount'].mean()

print(f"The mean 'fare_amount' for rides with 0.0 Haversine distance is {mean_fare_for_zero_distance:.2f}.")
```

Haversine distance is unusual and could indicate an issue with the data. In a typical scenario, even very short rides would incur some fare amount due to a minimum fare or other charges.

```python
R = 6371.0
costliest_ride = data[data['fare_amount'] == data['fare_amount'].max()].iloc[0]

# Extract the coordinates
pickup_lat = costliest_ride['pickup_latitude']
pickup_lon = costliest_ride['pickup_longitude']
dropoff_lat = costliest_ride['dropoff_latitude']
dropoff_lon = costliest_ride['dropoff_longitude']

# Function to calculate Haversine distance
def haversine_distance(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])
    
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    distance = R * c
    return distance

# Calculate the Haversine distance for the costliest ride
distance_costliest_ride = haversine_distance(pickup_lat, pickup_lon, dropoff_lat, dropoff_lon)
print(f"The Haversine distance between pickup and dropoff locations for the costliest ride is {distance_costliest_ride: .6f} kilometers.")
```

A very high 'fare_amount' for a relatively short distance might indicate an error in the dataset or an unusual ride scenario.  Such as checking for anomalies in the 'fare_amount' and 'distance' columns or reviewing the details of the costliest ride, can help you assess the data quality and identify potential issues.

```python
data['ride_week_day'] = data['pickup_datetime'].dt.dayofweek
data.dropna(inplace=True)
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
X = data[['passenger_count', 'distance', 'ride_week_day']]
y = data['fare_amount']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train different regression models
linear_reg = LinearRegression()
linear_reg.fit(X_train, y_train)

decision_tree_reg = DecisionTreeRegressor()
decision_tree_reg.fit(X_train, y_train)

random_forest_reg = RandomForestRegressor()
random_forest_reg.fit(X_train, y_train)

knn_reg = KNeighborsRegressor()
knn_reg.fit(X_train, y_train)

# Make predictions on the test data
linear_reg_predictions = linear_reg.predict(X_test)
decision_tree_predictions = decision_tree_reg.predict(X_test)
random_forest_predictions = random_forest_reg.predict(X_test)
knn_predictions = knn_reg.predict(X_test)

# Calculate R-squared values for each model
linear_reg_r2 = r2_score(y_test, linear_reg_predictions)
decision_tree_r2 = r2_score(y_test, decision_tree_predictions)
random_forest_r2 = r2_score(y_test, random_forest_predictions)
knn_r2 = r2_score(y_test, knn_predictions)

print("Linear Regression R-squared:", linear_reg_r2)
print("Decision Tree Regression R-squared:", decision_tree_r2)
print("Random Forest Regression R-squared:", random_forest_r2)
print("K-Nearest Neighbors Regression R-squared:", knn_r2)
```
